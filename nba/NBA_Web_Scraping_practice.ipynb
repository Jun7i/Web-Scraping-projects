```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
#import matplotlib.pyplot as plt
import numpy as np
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
table = soup.find_all(class_="full_table")
```

```python
head = soup.find(class_="thead")
column_names_raw = [head.text for item in head][0]
column_names_raw
```

```python
column_name_clean = column_names_raw.replace("\n",",").split(",")[2:-1]
column_name_clean
```

```python
players = []
for i in range(len(table)):
    player_=[]
    for td in table[i].find_all("td"):
        player_.append(td.text)
    players.append(player_)
df = pd.DataFrame(players, columns = column_name_clean).set_index("Player")
df.index = df.index.str.replace("*","")
```

```python
df
```

```python
df[['PTS']] = df[['PTS']].astype(float)
```

```python
lst = [i for i in df['Age']]
age_dict = {item:lst.count(item) for item in lst}
key_lst = [key for key, val in age_dict.items()]
val_lst = [val for key, val in age_dict.items()]
age_df = pd.DataFrame({'ages': key_lst,'count': val_lst})
age_df
```
