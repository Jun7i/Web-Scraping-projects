```python
# Flatten to chinese_df
chinese_df = pd.json_normalize(chinese_restaurants['businesses'])
chinese_df.head()
```

```python
# Select only necessary columns
chinese_df = chinese_df[['id', 'alias', 'name', 'review_count', 'rating']]
```

```python
chinese_reviews_df = pd.DataFrame()

for i in range(chinese_df.shape[0]):
  reviews = yelp_api.reviews_query(id = chinese_df.id[i])
  reviews_df = pd.json_normalize(reviews['reviews'])
  reviews_df['location_id'] = chinese_df['id'][i]
  chinese_reviews_df = chinese_reviews_df.append(reviews_df)
```

```python
# Select columns 'id', 'text', 'rating', 'time_created', 'location_id'
chinese_reviews_df = chinese_reviews_df[['id', 'text', 'rating', 'time_created', 'location_id']]
```

```python
# Do groupby to get mean rating. 
latest_reviews_agg = chinese_reviews_df.groupby('location_id', as_index=False).agg({'rating' : ['mean']})
```

```python
# Rename so the two column names are 'location_id' and 'mean_rating'
latest_reviews_agg.columns = ['location_id', 'mean_rating']
latest_reviews_agg
```

```python
# Join and name chinese_comp
chinese_comp = chinese_df.merge(latest_reviews_agg, 'outer', left_on='id', right_on='location_id')
```

```python
# Make still_good column
chinese_comp['still_good'] = np.where(chinese_comp['mean_rating'] >= chinese_comp['rating'], 'yes', 'no')
chinese_comp.head()
```
